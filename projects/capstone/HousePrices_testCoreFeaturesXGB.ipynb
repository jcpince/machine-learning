```python
import pandas as pd
import numpy as np
import matplotlib

import matplotlib.pyplot as plt

from HousePrices_utils import encode_by_SalePrice, add_sold_age, train_model, combine_features

from IPython import display

%config InlineBackend.figure_format = 'retina' #set 'png' here when working on notebook
%matplotlib inline

random_state = 67
nfolds = 5
```

```python
# Load the training dataset
train = pd.read_csv('input/train_full.csv').set_index('Id')
submit = pd.read_csv('input/test_full.csv').set_index('Id')
saleprices = np.log(train['SalePrice'])
whole_ds = pd.concat([train.drop('SalePrice', axis=1), submit])

# Encode the YrSold, MoSold into a single SoldMonths feature
whole_ds = add_sold_age(whole_ds)

# Removing TotalBsmtSF since it is highly correlated to 1stFlrSF and is the sum of the
# Basement areas
_ = whole_ds.pop('TotalBsmtSF')

# Encode the categorical features
whole_enc = encode_by_SalePrice(whole_ds, train)

train_X = whole_enc.ix[train.index]
```

```python
# Those are the features I beleive matter most in estimating the house
ALL_FEATURES = whole_enc.columns.tolist()
PRIMARY_FEATURES = ['GrLivArea', '1stFlrSF', 'TotRmsAbvGrd', 'FullBath', 'HalfBath', 'BsmtFullBath', 'GarageCars',
                 'PoolArea', 'SoldMonths', 'LotArea', 'Neighborhood', 'OverallQual', 'YearBuilt', 'Exterior1st',
                 'BsmtFinSF1']

# Secondary features
SECONDARY_FEATURES = list(set(ALL_FEATURES) - set(PRIMARY_FEATURES))

train_X_prim = train_X[PRIMARY_FEATURES].copy()
train_X_sec = train_X[SECONDARY_FEATURES].copy()
```

```python
from sklearn.model_selection import cross_val_score

def rmse_cv(model, X, Y, nfolds=nfolds):
    rmse= np.sqrt(-cross_val_score(model, X, Y,
                                   scoring="neg_mean_squared_error", cv = 5))
    return(rmse)
```

```python
import xgboost as xgb

dtrain = xgb.DMatrix(train_X_prim, label = saleprices)

#params = {"max_depth":2, "eta":0.05} # test_rmse around 0.142, train_rmse around 0.105
params = {"max_depth":3, "eta":0.03} # test_rmse around 0.148, train_rmse around 0.075
#clf = xgb.XGBRegressor(max_depth=6, learning_rate=0.003,
#    seed=random_state, n_estimators=20000, subsample=0.7)
cv_model = xgb.cv(params, dtrain,  num_boost_round=5000, early_stopping_rounds=100, metrics='rmse', nfold=5)
```

```python
cv_model.loc[150:,["test-rmse-mean", "train-rmse-mean"]].plot()
```

```python
depths = [2, 3, 4, 5, 6, 7, 8, 9]
cv_results = [rmse_cv(xgb.XGBRegressor(max_depth=depth, learning_rate=0.03, min_child_weight=3,
                    seed=random_state, n_estimators=1200, subsample=0.7), train_X_prim, saleprices).mean() for depth in depths]

print("Best rmse: %0.6f, best depth: %f (%0.6f)" % (
    np.array(cv_results).min(), depths[np.argmin(cv_results)], cv_results[np.argmin(cv_results)]))

cv_results = pd.Series(cv_results, index = depths)
cv_results.plot(title = "Validation - Just Do It")
plt.xlabel("child weight")
plt.ylabel("rmse")
```

```python
xgb_reg1 = xgb.XGBRegressor(max_depth=3, learning_rate=0.03, min_child_weight=3,
                    seed=random_state, n_estimators=1200, subsample=0.7).fit(train_X_prim, saleprices)
train_X_sec['prim_preds'] = xgb_reg1.predict(train_X_prim)
```

```python
dtrain = xgb.DMatrix(train_X_sec, label = saleprices)

#params = {"max_depth":2, "eta":0.05} # test_rmse around 0.142, train_rmse around 0.105
params = {"max_depth":3, "eta":0.03} # test_rmse around 0.148, train_rmse around 0.075
cv_model = xgb.cv(params, dtrain,  num_boost_round=5000, early_stopping_rounds=100, metrics='rmse', nfold=5)
```

```python
cv_model.loc[150:,["test-rmse-mean", "train-rmse-mean"]].plot()
```

```python
xgb_reg2 = xgb.XGBRegressor(max_depth=3, learning_rate=0.03, min_child_weight=3,
                    seed=random_state, n_estimators=240, subsample=0.7).fit(train_X_sec, saleprices)
```

```python
submit_X = whole_enc.ix[submit.index]
submit_X_prim = submit_X[PRIMARY_FEATURES].copy()
submit_X_sec = submit_X[SECONDARY_FEATURES].copy()

submit_X_sec['prim_preds'] = xgb_reg1.predict(submit_X_prim)

preds = xgb_reg2.predict(submit_X_sec)
```

```python
l1_submit_preds = pd.DataFrame(index=submit_X_sec.index)
l1_submit_preds['SalePrice'] = np.exp(preds)
l1_submit_preds.to_csv("submission_xgb_split.csv", index=True)
```
