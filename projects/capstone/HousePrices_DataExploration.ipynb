# Check features co-linearities

The data set is very rich in features but quite poor in records; 79 features for only 1460 records...

This might lead to models extracting the wrong sense out of the data.

The categorical data shall be encoded first and I'm using LabelEncoder class to avoid multiplicating the number of features.

Let's check how the features are related:

```python
categorical_features = whole_ds.dtypes[whole_ds.dtypes == "object"].index.tolist()
numeric_features = whole_ds.dtypes[whole_ds.dtypes != "object"].index

# Transfom the categorical features into integers to plot the heatmap
categorical_features = whole_ds.dtypes[whole_ds.dtypes == "object"].index.tolist()
encoded_cats = pd.DataFrame(index=whole_ds.index)
from sklearn import preprocessing
for feat in categorical_features:
    le = preprocessing.LabelEncoder()
    le.fit(whole_ds[feat].tolist())
    encoded_cats[feat] = le.transform(whole_ds[feat].tolist()) 

plt.figure(figsize=(24,24))
whole_with_encoded_cats = whole_ds[numeric_features].copy()
whole_with_encoded_cats = pd.concat([whole_with_encoded_cats, encoded_cats], axis=1)
corr = whole_with_encoded_cats.corr()
sns.set(font_scale=1.5)
sns.heatmap(corr, vmax=1, square=True)
```

## Analysis

As expected, there are a few features that are closely related such as YearBuilt, YearRemodAdd and GarageYrBlt since the garage and house are often built at the same time.

There is also a clear relationship between the TotalBsmtSF and 1stFlrSF since it is frequent to build a house with a basement on the whole root area of the house (so the 1st floor area).

Also, a strong relationship between the GarageArea and the GarageCars since we usually built a bigger garage when we expect to put more cars in it...

Exterior1st and Exterior2nd are closely related; probably because houses are built with materials that have a certain coherency. If Exterior1st is stone, there is probably a low chance that Exterior2nd was some vinyl.

Finally, we can observe that the EnclosedPorchto SaleCondition are mostly uncorrelated to the rest. Here the year of the sell might have an influence since the sale prices use to vary in time for a same property.

Out of this, I would assume that keeping TotalBsmtSF and 1stFlrSF would probably be a penalty to the linear models.

# Features impact on the Sale price:

Now, let's check the relations between all those features and the saleprice:

```python
train_encoded = whole_with_encoded_cats.ix[train.index]
train_encoded['SalePrice'] = train.SalePrice

corr = train_encoded.corr()[['SalePrice']]
corr.sort_values('SalePrice', inplace=True, ascending=False)
sns.set(font_scale=2)
plt.figure(figsize=(12,30))
sns.heatmap(corr, vmax=1, square=True)
```

## Analysis

The numerical features impact seems quite ok like the area based features but the categorical features seem really odd...
The relationship between Neighborhood and SalePrice is close to 0! But it shouldn't!!

Let's try to encode the features by their impact on the sale Price:

```python
def get_feature_order(ds, feature):
    medians = ds.groupby([feature])['SalePrice'].median()
    return sorted(medians.keys(), key=lambda x: medians[x])

def encode_by_SalePrice(ds_old, ds_with_price):
    ds = ds_old.copy()
    for feature in ds.columns.values:
            ftype = ds[feature].dtype
            if ftype == np.float64 or ftype == np.float32 or \
                    ftype == np.int64 or ftype == np.int32:
                    continue
            order = get_feature_order(ds_with_price, feature)
            class_mapping = {v: order.index(v) for v in order}
            ds[feature] = ds[feature].map(class_mapping)
    #assert(ds.isnull().sum().sum() == 0)
    return ds

whole_encoded = encode_by_SalePrice(whole_ds, train)
```

```python
plt.figure(figsize=(24,24))
corr = whole_encoded.corr()
sns.set(font_scale=1.5)
sns.heatmap(corr, vmax=1, square=True)

train_encoded = whole_encoded.ix[train.index]
train_encoded['SalePrice'] = train.SalePrice
corr = train_encoded.corr()[['SalePrice']]
corr.sort_values('SalePrice', inplace=True, ascending=False)
sns.set(font_scale=2)
plt.figure(figsize=(12,30))
sns.heatmap(corr, vmax=1, square=True)
```

## Analysis

This seems much better, Neighborhood for example has a much bigger impact on the sale price which is totally expected.

However, I feel odd that YrSold was less related to the SalePrice than MoSold... so I think we should consolidate both into a SoldSince in months.

We can also observe that few data are mostly uncorrelated to the rest of the features (the porch related features seem quite independant).

I would have expected the YrSold to have a much bigger impact on the sale price but 2006-2010 was probably a stable period in Ames.
