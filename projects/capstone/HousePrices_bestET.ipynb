```python
import pandas as pd
import numpy as np
import matplotlib

import matplotlib.pyplot as plt

from IPython import display

%config InlineBackend.figure_format = 'retina' #set 'png' here when working on notebook
%matplotlib inline

random_state = 67
nfolds = 5

# Load the training dataset
train = pd.read_csv('input/train_full.csv').set_index('Id')
submit = pd.read_csv('input/test_full.csv').set_index('Id')
saleprices = np.log(train['SalePrice'])
whole_ds = pd.concat([train.drop('SalePrice', axis=1), submit])

from HousePrices_utils import encode_by_SalePrice, add_sold_age, train_model

# Encode the YrSold, MoSold into a single SoldMonths feature
whole_ds = add_sold_age(whole_ds)

# Removing TotalBsmtSF since it is highly correlated to 1stFlrSF and is the sum of the
# Basement areas
_ = whole_ds.pop('TotalBsmtSF')

# Encode the categorical features
whole_enc = encode_by_SalePrice(whole_ds, train)

train_X = whole_enc.ix[train.index]
train_X = train_X[train_X.GrLivArea < 4000]
saleprices = saleprices.ix[train_X.index]
submit_X = whole_enc.ix[submit.index]
```

```python
from sklearn.model_selection import cross_val_score

def rmse_cv(model, X, Y, nfolds=nfolds):
    rmse= np.sqrt(-cross_val_score(model, X, Y,
                                   scoring="neg_mean_squared_error", cv = 5))
    return(rmse)
```

```python
from sklearn.ensemble import ExtraTreesRegressor

depths = [8, 9, 10, 11, 12, 13, 15, 17]
cv_results = [rmse_cv(ExtraTreesRegressor(max_depth=depth, criterion='mse',
            random_state=random_state, 
            n_estimators=500, max_features=8), train_X, saleprices).mean() for depth in depths]

print("Best rmse: %0.6f, best depth: %f (%0.6f)" % (
    np.array(cv_results).min(), depths[np.argmin(cv_results)], cv_results[np.argmin(cv_results)]))

cv_results = pd.Series(cv_results, index = depths)
cv_results.plot(title = "Validation - Just Do It")
plt.xlabel("child weight")
plt.ylabel("rmse")
```

```python
from threading import Lock

thread_lock = Lock()
submit_X = whole_enc.ix[submit.index]

l1_train_preds = pd.DataFrame(index=train_X.index)
l1_submit_preds = pd.DataFrame(index=submit_X.index)

clf = ExtraTreesRegressor(max_depth=17, criterion='mse', random_state=random_state, 
            n_estimators=500, max_features=8)

transformation='log'
train_model('SalePrice', clf, train_X, np.exp(saleprices), submit_X,
    l1_train_preds, l1_submit_preds, 5, float('inf'),
    thread_lock, random_state, transformation=transformation)
```

```python
l1_submit_preds.to_csv("submission_et.csv", index=True)
```

```python
def rmsecv_log(n_estimators, max_depth, min_samples_split, max_features):
    rmse = np.sqrt(-cross_val_score(ExtraTreesRegressor(criterion='mse',
                    n_estimators=int(n_estimators),
                    max_depth=int(max_depth),
                    min_samples_split=int(min_samples_split),
                    max_features=min(max_features, 0.999),
                    random_state=random_state),
                    train_X, saleprices, scoring="neg_mean_squared_error", cv = 5))
    return(1/rmse.mean())
```

```python
from sklearn.cross_validation import cross_val_score
from bayes_opt import BayesianOptimization

gp_params = {"alpha": 1e5}

etlogBO = BayesianOptimization(rmsecv_log, {'n_estimators': (10, 800),
                                     'max_depth': (4,20),
                                     'min_samples_split': (2, 25),
                                     'max_features': (0.1, 0.999)})
etlogBO.maximize(n_iter=100, **gp_params)
```

```python
print(etlogBO.res['max']['max_params'])
```
