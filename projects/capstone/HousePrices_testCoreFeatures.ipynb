```python
import pandas as pd
import numpy as np
import matplotlib

import matplotlib.pyplot as plt

from HousePrices_utils import encode_by_SalePrice, add_sold_age, train_model, combine_features

from IPython import display

%config InlineBackend.figure_format = 'retina' #set 'png' here when working on notebook
%matplotlib inline

random_state = 67
nfolds = 5
```

```python
# Load the training dataset
train = pd.read_csv('input/train_full.csv').set_index('Id')
submit = pd.read_csv('input/test_full.csv').set_index('Id')
saleprices = np.log(train['SalePrice'])
whole_ds = pd.concat([train.drop('SalePrice', axis=1), submit])

# Encode the YrSold, MoSold into a single SoldMonths feature
whole_ds = add_sold_age(whole_ds)

# Removing TotalBsmtSF since it is highly correlated to 1stFlrSF and is the sum of the
# Basement areas
_ = whole_ds.pop('TotalBsmtSF')

# Encode the categorical features
whole_enc = encode_by_SalePrice(whole_ds, train)

train_X = whole_enc.ix[train.index]
```

```python
# Those are the features I beleive matter most in estimating the house
ALL_FEATURES = whole_enc.columns.tolist()
PRIMARY_FEATURES = ['GrLivArea', '1stFlrSF', 'TotRmsAbvGrd', 'FullBath', 'HalfBath', 'BsmtFullBath', 'GarageCars',
                 'PoolArea', 'SoldMonths', 'LotArea', 'Neighborhood', 'OverallQual', 'YearBuilt', 'Exterior1st',
                 'BsmtFinSF1']

# Secondary features
SECONDARY_FEATURES = list(set(ALL_FEATURES) - set(PRIMARY_FEATURES))
```

```python
train_X_prim = train_X[PRIMARY_FEATURES].copy()
train_X_sec = train_X[SECONDARY_FEATURES].copy()
```

```python
from sklearn.model_selection import cross_val_score

def rmse_cv(model, X, Y, nfolds=nfolds):
    rmse= np.sqrt(-cross_val_score(model, X, Y,
                                   scoring="neg_mean_squared_error", cv = 5))
    return(rmse)
```

```python
from sklearn.linear_model import Lasso
alphas = [0.00001, 0.0001, 0.0005, 0.0009, 0.00095, 0.001]
cv_results = [rmse_cv(Lasso(alpha = alpha, random_state=random_state), train_X_prim, saleprices).mean() for alpha in alphas]

print("Best rmse: %0.6f, best alpha: %f (%0.6f)" % (
    np.array(cv_results).min(), alphas[np.argmin(cv_results)], cv_results[np.argmin(cv_results)]))

cv_results = pd.Series(cv_results, index = alphas)
cv_results.plot(title = "Validation - Just Do It")
plt.xlabel("alpha")
plt.ylabel("rmse")
```

```python
lasso = Lasso(alpha=0.001, random_state=random_state).fit(train_X_prim, saleprices)
train_X_sec['prim_preds'] = lasso.predict(train_X_prim)
```

```python
alphas = [0.0005, 0.0009, 0.00095, 0.001, 0.0015, 0.002]
cv_results = [rmse_cv(Lasso(alpha = alpha, random_state=random_state), train_X_sec, saleprices).mean() for alpha in alphas]

print("Best rmse: %0.6f, best alpha: %f (%0.6f)" % (
    np.array(cv_results).min(), alphas[np.argmin(cv_results)], cv_results[np.argmin(cv_results)]))

cv_results = pd.Series(cv_results, index = alphas)
cv_results.plot(title = "Validation - Just Do It")
plt.xlabel("alpha")
plt.ylabel("rmse")
```

```python
prim_error = saleprices - train_X_sec['prim_preds']
matplotlib.rcParams['figure.figsize'] = (24.0, 10.0)
prim_error.hist(bins=20)
print(prim_error.describe())

# Outliers
IQR = prim_error.quantile(.75) - prim_error.quantile(.25)
LOW_BOUND = prim_error.quantile(.25) - 1.5*IQR
HIGH_BOUND = prim_error.quantile(.75) + 1.5*IQR

inliers = prim_error[prim_error >= LOW_BOUND]
inliers = inliers[inliers <= HIGH_BOUND]
train_X_sec_inliers = train_X_sec.ix[inliers.index]
saleprices_inliers = saleprices.ix[inliers.index]

prim_error = saleprices_inliers - train_X_sec_inliers['prim_preds']
prim_error.hist(bins=20)
```

```python
train_X_prim_inliers = train_X_prim.ix[inliers.index].copy()
train_X_sec_inliers = train_X_sec.ix[inliers.index].copy()
saleprices_inliers = saleprices.ix[inliers.index]

alphas = [0.00001, 0.0001, 0.0005, 0.0009, 0.00095, 0.001]
cv_results = [rmse_cv(Lasso(alpha = alpha, random_state=random_state),
                      train_X_prim_inliers, saleprices_inliers).mean() for alpha in alphas]

print("Best rmse: %0.6f, best alpha: %f (%0.6f)" % (
    np.array(cv_results).min(), alphas[np.argmin(cv_results)], cv_results[np.argmin(cv_results)]))

cv_results = pd.Series(cv_results, index = alphas)
cv_results.plot(title = "Validation - Just Do It")
plt.xlabel("alpha")
plt.ylabel("rmse")
```

```python
lasso = Lasso(alpha=0.001, random_state=random_state).fit(train_X_prim_inliers, saleprices_inliers)
train_X_sec_inliers['prim_preds'] = lasso.predict(train_X_prim_inliers)
```

```python
alphas = [0.0005, 0.0009, 0.00095, 0.001, 0.0012, 0.0015, 0.002]
cv_results = [rmse_cv(Lasso(alpha = alpha, random_state=random_state),
                      train_X_sec_inliers, saleprices_inliers).mean() for alpha in alphas]

print("Best rmse: %0.6f, best alpha: %f (%0.6f)" % (
    np.array(cv_results).min(), alphas[np.argmin(cv_results)], cv_results[np.argmin(cv_results)]))

cv_results = pd.Series(cv_results, index = alphas)
cv_results.plot(title = "Validation - Just Do It")
plt.xlabel("alpha")
plt.ylabel("rmse")
```

```python
lasso1 = Lasso(alpha=0.001, random_state=random_state).fit(train_X_prim_inliers, saleprices_inliers)
lasso2 = Lasso(alpha=0.0012, random_state=random_state).fit(train_X_sec_inliers, saleprices_inliers)

submit_X = whole_enc.ix[submit.index]
submit_X_prim = submit_X[PRIMARY_FEATURES].copy()
submit_X_sec = submit_X[SECONDARY_FEATURES].copy()
submit_X_sec['prim_preds'] = lasso1.predict(submit_X_prim)
result = pd.DataFrame(index=submit_X_sec.index)
result['SalePrice'] = np.exp(lasso2.predict(submit_X_sec))
```
