```python
import pandas as pd
import numpy as np
import matplotlib

import matplotlib.pyplot as plt

from IPython import display

%config InlineBackend.figure_format = 'retina' #set 'png' here when working on notebook
%matplotlib inline

random_state = 67
nfolds = 5
```

```python
# Load the training dataset
train = pd.read_csv('input/train_full.csv').set_index('Id')
submit = pd.read_csv('input/test_full.csv').set_index('Id')
saleprices = np.log(train['SalePrice'])
whole_ds = pd.concat([train.drop('SalePrice', axis=1), submit])

from HousePrices_utils import encode_by_SalePrice, add_sold_age

# Encode the YrSold, MoSold into a single SoldMonths feature
whole_ds = add_sold_age(whole_ds)

# Removing TotalBsmtSF since it is highly correlated to 1stFlrSF and is the sum of the
# Basement areas
_ = whole_ds.pop('TotalBsmtSF')

# Encode the categorical features
whole_enc = encode_by_SalePrice(whole_ds, train)

train_X = whole_enc.ix[train.index]
train_X = train_X[train_X.GrLivArea < 4000]
saleprices = saleprices.ix[train_X.index]
```

```python
from sklearn.model_selection import cross_val_score

def rmse_cv(model, X, Y, nfolds=nfolds):
    rmse= np.sqrt(-cross_val_score(model, X, Y,
                                   scoring="neg_mean_squared_error", cv = 5))
    return(rmse)
```

```python
from sklearn.linear_model import Lasso
alphas = [0.0009, 0.00095, 0.001, 0.0015, 0.0017, 0.002, 0.003]
cv_results = [rmse_cv(Lasso(alpha = alpha, random_state=random_state
                            ), train_X, saleprices).mean() for alpha in alphas]

#sklearn.linear_model.Lasso(alpha=1.0, fit_intercept=True, normalize=False, precompute=False, copy_X=True,
#                           max_iter=1000, tol=0.0001, warm_start=False, positive=False, random_state=None, selection='cyclic')

print("Best rmse: %0.6f, best alpha: %f (%0.6f)" % (
    np.array(cv_results).min(), alphas[np.argmin(cv_results)], cv_results[np.argmin(cv_results)]))

cv_results = pd.Series(cv_results, index = alphas)
cv_results.plot(title = "Validation - Just Do It")
plt.xlabel("alpha")
plt.ylabel("rmse")
```

```python
from sklearn.preprocessing import StandardScaler, RobustScaler
standard_scaler = StandardScaler()
Xtr_s = standard_scaler.fit_transform(train_X)

robust_scaler = RobustScaler()
Xtr_r = robust_scaler.fit_transform(train_X)
```
