```python
import pandas as pd
import numpy as np
import matplotlib

import matplotlib.pyplot as plt

from IPython import display

%config InlineBackend.figure_format = 'retina' #set 'png' here when working on notebook
%matplotlib inline

random_state = 67
nfolds = 5

# Load the training dataset
train = pd.read_csv('input/train_full.csv').set_index('Id')
submit = pd.read_csv('input/test_full.csv').set_index('Id')
saleprices = train['SalePrice']
whole_ds = pd.concat([train.drop('SalePrice', axis=1), submit])

from HousePrices_utils import encode_by_SalePrice, add_sold_age, train_model

# Encode the YrSold, MoSold into a single SoldMonths feature
whole_ds = add_sold_age(whole_ds)

# Removing TotalBsmtSF since it is highly correlated to 1stFlrSF and is the sum of the
# Basement areas
_ = whole_ds.pop('TotalBsmtSF')

# Encode the categorical features
whole_enc = encode_by_SalePrice(whole_ds, train)

train_X = whole_enc.ix[train.index]
train_X = train_X[train_X.GrLivArea < 4000]
saleprices = saleprices.ix[train_X.index]
submit_X = whole_enc.ix[submit.index]
```

```python
from threading import Lock

thread_lock = Lock()

l1_train_preds = pd.DataFrame(index=train_X.index)
l1_submit_preds = pd.DataFrame(index=submit_X.index)

l1_best_algos = ['et_combined_boruta', 'gb', 'lasso_combined_boruta',
                 'linreg', 'rf_combined_boruta',
                 'ridge_combined_boruta', 'xgb_combined_boruta']

l1_all_algos = l1_best_algos + ['et', 'gb_combined_boruta', 'lasso',
                'rf',
                'ridge', 'xgb']

for algo in l1_best_algos:
    tmp = pd.read_csv('train_' + algo + '.csv').set_index('Id')
    l1_train_preds[algo] = tmp
    tmp = pd.read_csv('submission_' + algo + '.csv').set_index('Id')
    l1_submit_preds[algo] = tmp

#submission_et_combined_boruta 0.1345
#submission_et 0.1411
#submission_gb_combined_boruta 0.1266
#submission_gb 0.1253
#submission_inliers 0.1367
#submission_lasso2 0.1343
#submission_lasso 0.1324
#submission_lasso_combined_boruta 0.1266
#submission_linreg_combined_boruta 0.1713
#submission_linreg 0.1340
#submission_rf_combined_boruta 0.1379
#submission_rf 0.1396
#submission_ridge_combined_boruta 0.1207
#submission_ridge 0.1331
#submission_xgb_combined_boruta 0.1260
#submission_xgb 0.1278
#submission_xgb_split 0.1390
```

```python
from HousePrices_utils import get_layer_average, rmsdl
l1_mean = get_layer_average(l1_train_preds)
for algo in l1_train_preds:
    print('RMSD train error for %s: %f' % (algo, rmsdl(saleprices, l1_train_preds[algo])))
print('RMSD train error: %f' % rmsdl(saleprices, l1_mean))

l1_mean = pd.DataFrame(index=l1_submit_preds.index)
l1_mean['SalePrice'] = get_layer_average(l1_submit_preds)
l1_mean.to_csv("submission_l2_mean.csv", index=True)
print('LB score: 0.1223')
```

```python
from sklearn.model_selection import cross_val_score

def rmse_cv(model, X, Y, nfolds=nfolds):
    rmse= np.sqrt(-cross_val_score(model, X, Y,
                                   scoring="neg_mean_squared_error", cv = 5))
    return(rmse)
```

```python
from sklearn.linear_model import Ridge
alphas = [0.1, 0.25, 0.5, 0.75, 1, 5]
cv_results = [rmse_cv(Ridge(alpha = alpha, random_state=random_state),
                      np.log(l1_train_preds), np.log(saleprices)).mean() for alpha in alphas]

print("Best rmse: %0.6f, best alpha: %f (%0.6f)" % (
    np.array(cv_results).min(), alphas[np.argmin(cv_results)], cv_results[np.argmin(cv_results)]))

cv_results = pd.Series(cv_results, index = alphas)
cv_results.plot(title = "Validation - Just Do It")
plt.xlabel("alpha")
plt.ylabel("rmse")
```

```python
from threading import Lock

thread_lock = Lock()

l2_train_preds = pd.DataFrame(index=train_X.index)
l2_submit_preds = pd.DataFrame(index=submit_X.index)

clf = Ridge(alpha=1.0, random_state=random_state)

transformation='log'
train_model('SalePrice', clf, np.log(l1_train_preds), saleprices, np.log(l1_submit_preds),
    l2_train_preds, l2_submit_preds, 5, float('inf'),
    thread_lock, random_state, transformation=transformation)
```

```python
l2_train_preds.to_csv("train_l2_ridge.csv", index=True)
l2_submit_preds.to_csv("submission_l2_ridge.csv", index=True)
```
